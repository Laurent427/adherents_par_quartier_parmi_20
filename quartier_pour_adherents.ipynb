```python
# Importation des packages python
import numpy as np
import pandas as pd
```

TEST AVEC FICHIER ADRESSES POUR LEQUEL IL FAUT AJOUTER LE QUARTIER

```python
df_test = pd.read_csv('test_adherent.csv', sep=";")
display(df_test.head(10))
```

PREPARATION DES DONNEES :
- supprimer les caractètes accentués
- supprimer les ponctuations
- remplacer les abréviations par les mots entiers : BD, BLVD en BOULEVARD, puis ST en SAINT, etc
- mettre ne majuscule pour être au même format que le fichier de metropole Toulouse

```python

from unidecode import unidecode

# Pour la fonction suppression de la ponctuation et remplacement des abbréviations par les nots entiers
suppr_abbrev = {
      "." : " ", "," : " ", ":" : " ", "-" : " ", "/" : " ", " / " : " ", "-" : " ", "*" : " ", "'" : " ",
      "-" : " ",
      "\r" : " ", 
      "\n" : " ", # suppression des sauts à la ligne
      # "à": "a","â": "a","é": "e","è": "e","ê": "e","ë": "e", "î": "i","ï": "i", "ô": "o","ö": "o", "ù": "u","û": "u","ü": "u", "ç": "c", # fait avec methode unicode importée
      " blvd ": " boulevard "," bd ": " boulevard "," all ": " allee "," av ": " avenue "," ch ": " chemin "," qt ": " quartier ",
      " pl ": " place "," imp ": " impasse "," rte ": "route",
      " bis " : "b ", "bis " : "b ", " ter " : "t ", "ter " : "t ",
      " st " : " saint ", " ste ": " sainte ", " dr " : " docteur ", " pr " : " professeur "
    }
# Fonction de suppression de la ponctuation et remplacement des abbréviations par les nots entiers
def replace_abbrev(text):
  for abbr, word in suppr_abbrev.items():
    if abbr in text.lower():
      text = text.lower().replace(abbr, word)
  return text

# Suppression des caractères accentués 
df_test['Adresse_extenso'] = df_test['Adresse'].apply(unidecode)

# Opération de suppression de la ponctuation et remplacement des abbréviations par les nots entiers
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].apply(replace_abbrev)

# Suppresion de tous les caractères situés en fin du libellé d'adresse dès que l'on trouve Appt, Bat, Esc, Entrée, ... à compléter au besoin
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r"( apt | appt | appartement ).*$", "", regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r"( bat | batiment ).*$", "", regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r"( esc | escalier ).*$", "", regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r"( ent | entree ).*$", "", regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r' porte .*$', '', regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r'( res | residence ).*$', '', regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r' bal .*$', '', regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r' maison .*$', '', regex=True, case=False)
df_test["Adresse_extenso"] = df_test["Adresse_extenso"].str.replace(r'( les villa| villa).*$', '', regex=True, case=False)

display(df_test.head(10))
```

```python
# Formatage si besoin selon le fichier fourni pour avoir l'adresse complète dans une colonne
# et transformation texte en MAJUSCULES pour péraparer à la recherche
df_test['adresse'] = df_test['Adresse_extenso'].str.upper() + ' ' + df_test['Code postal'].astype(str) + ' ' + df_test['Commune'].str.upper()

# Suppression des espaces en trop : on ne garde qu'un seul espace entre 2 mots
df_test['adresse'] = df_test['adresse'].apply(lambda x: ' '.join(x.split()))

display(df_test.head(10))
```

```python
## Ajout quartier dans df_test par jointure avec df_rues_quartiers sur colonnes 'adresse'

# Chargement fichier csv créé grace aux données metropole Toulouse
df_rues_quartiers = pd.read_csv('numero-de-rue-quartier.csv')

# Fusionnez les DataFrames sur la colonne "adresse"
df_merged = pd.merge(df_test, df_rues_quartiers, on='adresse', how='left')

# Sélectionnez uniquement les colonnes de df1 et la colonne "quartier"
df_resultat = df_merged[['Adherent', 'adresse', 'quartier']]
print(type(df_resultat['quartier']) )

# Contrôle
display(df_resultat.head(10))
```

```python
# Création du fichier csv pour usage externe :-)
df_resultat.to_csv('resultats.csv')
```
